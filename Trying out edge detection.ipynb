```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv.imread("D:/User/Data_science/geek_for_geeks/Image_processing/images/book_flowers_dataScience.jpg")
```

```python
img.shape
```

```python
img=cv.resize(img,(480,640))
```

```python
plt.imshow(img)
```

```python
img=cv.GaussianBlur(img,(5,5),0)
```

```python
plt.imshow(img)
```

```python
img=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(img)
```

```python
laplacian=cv.Laplacian(img,cv.CV_64F)
# sobel=cv.Sobel(img,cv.CV_64F,-1,1)
```

```python
sobel_x=cv.Sobel(img,cv.CV_64F,1,0)
```

```python
sobel_y=cv.Sobel(img,cv.CV_64F,0,1)
```

```python
canny=cv.Canny(img,100,150)
```

```python
cv.imshow('Sobel_x',sobel_x)
cv.imshow('Sobel_y',sobel_y)
cv.imshow('Laplase',laplacian)
cv.imshow('Canny',canny)

key=cv.waitKey(0)
if key==32:
    cv.destroyAllWindows()
```
